```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configure plot style
sns.set(style="whitegrid")
```

```python
data = pd.read_csv('../data/raw/book_data_raw.csv')

data.head()

```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size for the plot
sns.histplot(data['price'], kde=True, bins=20)  # Plot the distribution of book prices with a KDE
plt.title('Distribution of Book Prices')  # Set the title of the plot
plt.xlabel('Price (USD)')  # Set the x-axis label
plt.ylabel('Frequency')  # Set the y-axis label
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size for the plot
sns.boxplot(x=data['price'])  # Create a boxplot of book prices
plt.title('Boxplot of Book Prices')  # Set the title of the plot
plt.xlabel('Price (USD)')  # Set the x-axis label
plt.show()
```

```python
# Top 10 Most Expensive Books
top_10_expensive_books = data.nlargest(10, 'price')  # Get the top 10 most expensive books

plt.figure(figsize=(12, 8))  # Set the figure size for the plot
sns.barplot(x='price', y='title', data=top_10_expensive_books)
```
